### Nans
Nan values can be extremely problematic when they sneak into your training data. For NLP problems, if you forget to remove nans and include them in you training sequence it will cause your model to crash at the training stage. For that reason, you are encouraged to remove nans from your data when possible.

Pandas offer's a method that [drops nan](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) values from your dataframe. Take note of the arguments: axis, how, and inplace. Read the documentation and see what those arguments do. Another helpful method is the [drop_duplicates](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html#pandas.DataFrame.drop_duplicates) method.

### Apply and lambdas

Are you feeling more comfortable with pandas yet? If not, please message the instructors for more clarification.

What makes pandas great is that you can vectorize many operations. Meaning that you can make changes to whole columns or rows with out writing for loops every time.

```python
#Let's make a new column called start year. Use apply to capture the start year
imdb_df['start_year'] = imdb_df['year'].apply(lambda x : int(x[1:5]))
imdb_df.head(5)
```

If we want to populate the start year column with the first year in the year column, it wouldn't be easy. The year column is a vector of strings. So we need to use [apply](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html) which will allow us to apply a function we write to modify all of the values in the columns.

As rule of thumb, the apply method is slow. If there is a pandas method that already has the vectorized operation, use that instead. For particular modifications, use apply. The lambda function is an anonymous function which tells apply what to do for each element in the column your are using apply for.

```python
#This function tokenizes a string. We will use this for our apply
def tokenize(string):
  tokens = word_tokenize(string)
  return [token for token in tokens if len(token) > 1]

print(tokenize("My name is Shaif"))
print(tokenize("(1959-1962)"))
```

```python
import torch
import torch.nn as nn

class LinearNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearNN, self).__init__()
        self.fc = nn.Linear(input_size, output_size)

    def forward(self, x):
        return self.fc(x)

input_size = X_train_scaled.shape[1]
output_size = 1  # Since you're predicting a single value
model = LinearNN(input_size, output_size)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
num_epochs = 5

for epoch in range(num_epochs):
    inputs = torch.tensor(X_train_scaled, dtype=torch.float32)
    targets = torch.tensor(Y_train.values, dtype=torch.float32).view(-1, 1)

    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```
